```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import json
import statsmodels.api as sam
from matplotlib.cm import ScalarMappable
```

```python
df = pd.read_csv("./data/multi_sentiments.csv")
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
#get the average polarity 12H and subjectivity
pol = df.copy()
pol['datetime'] = pd.to_datetime(pol['datetime'])
pol = pol.set_index('datetime')
pol = pol.resample('12H').mean()
plt.figure(figsize=(12,6));
sb.lineplot(x=pol.index,y=pol['polarity'],label='polarity')
sb.lineplot(x=pol.index,y=pol['subjectivity'],label='subjectivity')
```

```python
data = df.copy()
data['datetime'] = pd.to_datetime(data['datetime'])
data= data.set_index('datetime')
OLS = data.resample('12H').mean()
OLS = OLS['2020-04':]
x = OLS.index.day+OLS.index.month
m1,b = np.polyfit(x,OLS['polarity'],1)
m2,b2 = np.polyfit(x,OLS['subjectivity'],1)
```

```python
plt.figure(figsize=(10,6));
sb.scatterplot(x=OLS.index,y=OLS['polarity']);
sb.lineplot(x=OLS.index,y=x*m1+b,color = 'red');
plt.title('Polarity line');
plt.ylabel('polarity');
plt.xticks(rotation=70);
```

```python
plt.figure(figsize=(10,6));
sb.scatterplot(x=OLS.index,y=OLS['subjectivity']);
sb.lineplot(x=OLS.index,y=x*m2+b2,color='red');
plt.title('Subjectivity line');
plt.ylabel('subjectivity');
plt.xticks(rotation=70);
```

### A slight change in polarity from less than 0.12 to above 0.12 and some upward move in subjectivity from around 0.35 to 0.36

```python
#count of emotions each tweets contains

for i in range(0,len(df)):
    tone = df['document_tone'][i].replace("'",'"')
    tone = json.loads(tone)
    count = len(tone['tones'])
    df.loc[i,'tone_count']=count
df.head()
```

```python
#function to calculate sentiment percentage change

def sentiment_trend(sentiment):
    data = df.copy()
    data['datetime']= pd.to_datetime(data['datetime'])
    data = data.set_index('datetime')
    all_tweets = pd.DataFrame()
    all_tweets['tweet_count'] = data['text'].resample('12H').count()
    emo_count = data.query(f"{sentiment}!=0")
    all_tweets['emo_count'] = emo_count[sentiment].resample('12H').sum()
    all_tweets['sentiment_pct'] = all_tweets['emo_count']/all_tweets['tweet_count']
    return all_tweets

```

```python
emo = data.columns[-8:-1]
emo
```

```python
sadness = sentiment_trend('sadness')
tentative = sentiment_trend('tentative')
analytical = sentiment_trend('analytical')
confident = sentiment_trend('confident')
assertive = sentiment_trend('joy')
fear = sentiment_trend('fear')
anger = sentiment_trend('anger')
```

```python
plt.figure(figsize=(15,8))

for i in emo[:]:
    sb.lineplot(x=eval(i).index,y=eval(i).sentiment_pct,label = i)
plt.title('Sentiment % change')
plt.xticks(rotation=70);
```

```python
#sentiment density trend

sent_dense = pd.DataFrame()
data = df.copy()
data['datetime'] = pd.to_datetime(data['datetime'])
data= data.set_index('datetime')
sent_dense['tweet_count'] = data['text'].resample('12H').count()   # count of tweets every 12 h
sent_dense['emo_count'] = data['tone_count'].resample('12H').sum()    # number of emotions in each tweet
sent_dense['density_pct'] = sent_dense['emo_count']/sent_dense['tweet_count']  #sentiment sensity
sent_dense['pct_change'] = sent_dense['density_pct'].pct_change()
sent_dense.head()
```

```python
#plot the sentiment density
avg = sent_dense['density_pct'].mean()
plt.figure(figsize=(15,8));
sb.lineplot(x=sent_dense.index,y=sent_dense.density_pct,label='Sentiment Density')
sb.lineplot(x=sent_dense.index,y=np.ones((66))*avg,label='Average density')
plt.title("Sentiment Density trend")
```

```python
sent_dense.dropna(inplace=True)
data_color = np.abs(sent_dense['pct_change'])
data_color = [x /data_color.max() for x in data_color]
# data_color = sent_dense['pct_change']
fig, ax = plt.subplots(figsize=(15, 4));

my_cmap = plt.cm.get_cmap('GnBu')
colors = my_cmap(data_color)
rects = ax.bar(sent_dense.index, sent_dense['pct_change'], color=colors);
sm = ScalarMappable(cmap=my_cmap, norm=plt.Normalize(min(data_color),max(data_color)));
sm.set_array([]);

cbar = plt.colorbar(sm);
cbar.set_label('Color', rotation=270,labelpad=25);
plt.title("% change in Sentiment Density")
plt.xticks(rotation=70);
```

```python
def sentiment_polarity(emo):
    sent_pol = pd.DataFrame()
    for i in emo:
        d = pd.DataFrame()
        d[[f'{i}'+'_polairty',f'{i}'+'_subjectivity']] = data.query("{}==1".format(i))[['polarity','subjectivity']]
        sent_pol = pd.concat([sent_pol,d],axis=0)
        #     sent_pol = sent_pol.resample('12H').mean()
    return sent_pol
    
    
```

```python
sent_pol = sentiment_polarity(emo)
sent_pol = sent_pol.resample('12H').mean()
sent_pol
```
